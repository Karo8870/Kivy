```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from torch.nn.functional import cosine_similarity
import torch
from tqdm import tqdm

model = SentenceTransformer('all-MiniLM-L6-v2')


def replace_uncommon_keys_with_closest(series, threshold):
    value_counts = series.value_counts()

    common_keys = value_counts[value_counts > threshold].index
    uncommon_keys = value_counts[value_counts <= threshold].index

    common_key_embeddings = model.encode(common_keys, convert_to_tensor=True)

    key_to_closest_common = {}
    for key in uncommon_keys:
        key_embedding = model.encode(key, convert_to_tensor=True)

        similarities = cosine_similarity(key_embedding.unsqueeze(0), common_key_embeddings)

        closest_key_index = torch.argmax(similarities)
        key_to_closest_common[key] = common_keys[closest_key_index.item()]

    modified_series = series.replace(key_to_closest_common)

    return modified_series


def replace_uncommon_keys_list_with_closest(series_of_lists, threshold):
    # Flatten the lists to get all individual strings
    all_strings = [item for sublist in series_of_lists for item in sublist]
    value_counts = pd.Series(all_strings).value_counts()

    # Identify common and uncommon keys
    common_keys = value_counts[value_counts > threshold].index
    uncommon_keys = value_counts[value_counts <= threshold].index

    # Encode common keys
    common_key_embeddings = model.encode(common_keys, convert_to_tensor=True)

    # Map uncommon keys to the closest common key
    key_to_closest_common = {}

    for key in tqdm(uncommon_keys, desc="Processing keys"):
        key_embedding = model.encode(key, convert_to_tensor=True)

        similarities = cosine_similarity(key_embedding.unsqueeze(0), common_key_embeddings)
        closest_key_index = torch.argmax(torch.tensor(similarities))
        key_to_closest_common[key] = common_keys[closest_key_index.item()]

    # Replace uncommon keys in each list of strings
    def replace_keys_in_list(string_list):
        return [key_to_closest_common.get(key, key) for key in string_list]

    # Apply the replacement to the series of lists
    modified_series = series_of_lists.apply(replace_keys_in_list)

    return modified_series
```

```python
df['category'] = replace_uncommon_keys_with_closest(df['category'], 5)
df['tags'] = replace_uncommon_keys_list_with_closest(df['tags'].map(json.loads), 5)
df['ingredients'] = replace_uncommon_keys_list_with_closest(df['ingredients'].map(json.loads), 5)
```
